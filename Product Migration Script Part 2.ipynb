```python
import pandas as pd
import numpy as np
```

```python
### ACTION ### INPUT YOUR FILE PATH HERE
data = pd.read_excel(r'C:\Users\aizhang\OneDrive - ANSYS, Inc\Desktop\Ad-Hoc Tasks\Chase\Migration Automation\output1.xlsx', sheet_name= 'Migration-renewal Quantity')
data.shape
```

```python
result = []
for index, row in data.iterrows(): 
    path = row['Migration Path'].split(" + ") # you need to change the spliter if you used something else other than "+" (e.g. if you used "," or ";")
    if type(row['Migration License Quantity']) == str: 
        q = row['Migration License Quantity'].split(",")
    else: 
        q = row['Migration License Quantity']

    d = {}
    for i in range(len(path)):
        if type(q) == int:
            d[path[i]] = q
        else:
            d[path[i]] = q[i]
    
    result.append(d)

result
```

```python
### ACTION ### INPUT THE NEW PACKAGE'S PRICE TO CALCULATE THE NEW ACV AFTER MIGRATION FOR ACV IMPACT/COMPARISON ANALYSIS
price_d = {'Enterprise':97400, 'Premium Design':46000, 'Premium Sensor':46000, 'Premium Visualization': 64800, 'Pro':16800, 'Enterprise Solver': 58400, 'Premium TBD':46000}
acv=[]

for c in result:
    cur_val = 0
    for key in c.keys():
        price = price_d[key]
        q = c[key]
        value = price*int(q)
        cur_val+=value
    acv.append(cur_val)
```

```python
print(acv)
```

```python
data['ACV2'] = acv
data.shape
```

```python
result3 = []
for index, row in data.iterrows(): 
    path = row['Upsell Migration Path'].split(" + ")
    if type(row['Upsell Migration Quantity']) == str: 
        q = row['Upsell Migration Quantity'].split(",")
    else: 
        q = row['Upsell Migration Quantity']

    d = {}
    for i in range(len(path)):
        if type(q) == int:
            d[path[i]] = q
        else:
            d[path[i]] = q[i]
    
    result3.append(d)

result3
```

```python
### ACTION ### INPUT THE NEW PACKAGE'S PRICE TO CALCULATE THE NEW ACV AFTER MIGRATION WITH UPSELL FOR ACV IMPACT/COMPARISON ANALYSIS
price_d = {'Enterprise':97400, 'Premium Design':46000, 'Premium Sensor':46000, 'Premium Visualization': 64800, 'Pro':16800, 'Enterprise Solver': 58400, 'Premium TBD':46000,'-':0}
acv3=[]

for c in result3:
    cur_val = 0
    for key in c.keys():
        price = price_d[key]
        q = c[key]
        value = price*int(q)
        cur_val+=int(value)
    acv3.append(cur_val)
```

```python
print(acv3)
```

```python
data['ACV3'] = acv3
data.shape
```

```python
data['ACV3'] = data.apply(lambda row: row['ACV2'] if row['ACV3'] == 0 else row['ACV3'], axis=1)
```

```python
data['ACV2 - ACV1 Difference'] = data['ACV2'] - data['ACV1']
data['ACV3 - ACV1 Difference'] = data['ACV3'] - data['ACV1']
data['ACV3 - ACV2 Difference'] = data['ACV3'] - data['ACV2']
data['Degrees ACV2 - ACV1'] = data['ACV2 - ACV1 Difference']/data['ACV1']
```
